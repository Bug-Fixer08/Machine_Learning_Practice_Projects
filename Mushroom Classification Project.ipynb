```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['class'],axis=1)
y=df[['class']]
```

Data Preprocessing
Creating x and y values for Predicting the Model

Using Machine Learning to Build a

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
```

Importing necessary tools for Machine Learning

```python
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
    dtc=DecisionTreeClassifier()
    dtc.fit(x_train,y_train)
    pred=dtc.predict(x_test)
    acc=accuracy_score(y_test,pred)
    print(acc*100,'at_random_state',i)
```

```python
maxAccu=0

maxRS=0

for i in range (1,200):
    x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.25,random_state =i)
    

    dtc.fit(x_train,y_train)
    pred = dtc.predict(x_test)
    acc=accuracy_score(y_test,pred)


    if acc>maxAccu:
        maxAccu=acc
        maxRS=i

print("Best accuracy is ", maxAccu," on Random state" ,maxRS)
```

Selecting Random State no 1 for 100% Accuracy

# Model Building

```python
model=[LogisticRegression(),SVC(),DecisionTreeClassifier(),MultinomialNB()]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score in',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

Here we can see the accuracy score for differt models

                                 Accuaracy Score Cheak
Decision Tree Classifier         1.0
Logistic Regression              0.9497784342688331 
SVC                              0.9886755292959133
MultinomiaNB                     0.8030526834071886

```python
for m in model:
    score=cross_val_score(m,x,y,cv=5)
    print(m,score)
    print(m,score.mean())
    print('\n')
```

So After Cheaking the Cross Validation Score We can now see that DecisionTreeClassifier performns well with Mushroom Dataset.
